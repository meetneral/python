```python
import pandas as pd
```

```python
df=pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
```

```python
import matplotlib.pyplot as plt
import numpy as np
total_states=np.arange(len(df['state_name']))
```

```python
total_states
```

```python
max(df['positive'])
```

```python
from matplotlib.pyplot import figure
figure(num=None ,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k',alpha=0.5)
plt.barh(total_states,df['positive'],align='center',color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['positive'])+10)
plt.xlabel('Positive number of cases')
plt.title('Corona virus cases')

plt.show()
```

```python
from matplotlib.pyplot import figure
figure(num=None ,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k',alpha=0.5)
plt.barh(total_states,df['death'],align='center',color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['death'])+10)
plt.xlabel('Positive number of cases')
plt.title('Corona virus cases')

plt.show()
```

```python
from matplotlib.pyplot import figure
figure(num=None ,figsize=(9,6),dpi=80,facecolor='w',edgecolor='k',alpha=0.5)
plt.barh(total_states,df['new_active'],align='center',color=(1,0,0),edgecolor=(0.5,0.2,0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1,max(df['new_active'])+10)
plt.xlabel('Positive number of cases')
plt.title('Corona virus cases')

plt.show()
```
